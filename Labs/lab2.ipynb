```python
#1 append()
my_list = [1, 2, 3]
my_list.append(4)
print(f"After append(4): {my_list}")
```

```python
#2 clear()
my_list = [1, 2, 3]
my_list.clear()
print(f"After clear(): {my_list}")
```

```python
#3 copy() -
my_list = [1, 2, 3]
copied_list = my_list.copy()
print(f"Original list: {my_list}, Copied list: {copied_list}")
```

```python
#4 count() -
my_list = [1, 2, 2, 3, 2]
count_list=my_list.count(2)
print(f"Count of 2 in the list :{count_list}")
```

```python
#5 extend()
my_list = [1, 2, 3]
another_list = [4, 5]
my_list.extend(another_list)
print(f"After extend([4, 5]): {my_list}")
```

```python
#6 index() -
my_list = [1, 2, 3, 2]
index_of_2 = my_list.index(2)
print(f"Index of the first 2: {index_of_2}")
```

```python
#7 insert() -
my_list = [1, 3, 4]
my_list.insert(1, 2) # insert 2 at index 1
print(f"After insert(1, 2): {my_list}")
```

```python
#Bonus Task
fruits = ["apple", "banana", "orange", "apple", "mango", "banana"]
seen = set()
repeats = set()
for fruit in fruits:
    if fruit in seen:
        repeats.add(fruit)
    else:
        seen.add(fruit)
print(list(repeats))

```

```python
#8 pop() -
my_list = [1, 2, 3, 4]
popped_element = my_list.pop(2)
print(f"After pop: {my_list}")
```

```python
#9 remove()
my_list = [1, 2, 3, 4]
my_list.remove(4)
print(f"After remove(2): {my_list}")
```

```python
#10 reverse()
my_list = [1, 2, 3, 4]
my_list.reverse()
print(f"After reverse(): {my_list}")
```

```python
#11 sort()
my_list = [4, 2, 1, 3]
my_list.sort()
print(f"After sort(): {my_list}")
```
