```python
#Factorial
def factorial(num):
    if num == 0:
        return 1
    else:
        result = 1
        for i in range(1, num + 1):
            result *= i
        return result

num = int(input("Enter  number: "))
factorial(num)
```

```python
def countdigitsandsum(n):
    n = abs(n)
    digitcount = 0
    digitsum = 0
    while n > 0:
        digit = n % 10
        digitsum += digit
        digitcount += 1
        n //= 10
    return digitcount, digitsum
number = int(input("Enter a number: "))
count, total = countdigitsandsum(number)
print(f"Number of digits: ",count)
print(f"Sum of digits:",total)
```
