```python
def evaluate_tic_tac_toe(board):


    # Check rows for winner
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2]:
            if board[row][0] == 'x':
                return 10
            elif board[row][0] == 'o':
                return -10

    # Check columns for winner
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col]:
            if board[0][col] == 'x':
                return 10
            elif board[0][col] == 'o':
                return -10

    # Check diagonals for winner
    if board[0][0] == board[1][1] == board[2][2]:
        if board[0][0] == 'x':
            return 10
        elif board[0][0] == 'o':
            return -10

    if board[0][2] == board[1][1] == board[2][0]:
        if board[0][2] == 'x':
            return 10
        elif board[0][2] == 'o':
            return -10

    # No winner - draw or game still in progress
    return 0

def print_board(board):
    """Helper function to display the board"""
    print("Current Board:")
    for i in range(3):
        print(" | ".join(board[i]))
        if i < 2:
            print("---------")
    print()

# Sample boards to test the evaluation function
def test_evaluation_function():

    # Test case 1: X wins (row)
    board1 = [
        ['x', 'x', 'x'],
        ['o', 'o', '_'],
        ['_', '_', '_']
    ]

    # Test case 2: O wins (column)
    board2 = [
        ['x', 'o', '_'],
        ['x', 'o', '_'],
        ['_', 'o', '_']
    ]

    # Test case 3: X wins (diagonal)
    board3 = [
        ['x', 'o', '_'],
        ['o', 'x', '_'],
        ['_', '_', 'x']
    ]

    # Test case 4: Draw
    board4 = [
        ['x', 'o', 'x'],
        ['o', 'x', 'o'],
        ['o', 'x', 'o']
    ]

    # Test case 5: Game in progress
    board5 = [
        ['x', '_', '_'],
        ['_', 'o', '_'],
        ['_', '_', '_']
    ]

    # Test all boards
    test_boards = [board1, board2, board3, board4, board5]

    for i, board in enumerate(test_boards, 1):
        print(f"Test Case {i}:")
        print_board(board)
        score = evaluate_tic_tac_toe(board)
        print(f"Evaluation Score: {score}")

        if score == 10:
            print("Result: X wins!")
        elif score == -10:
            print("Result: O wins!")
        else:
            print("Result: Draw or game in progress")

        print("-" * 30)
        print()

# Interactive version to test custom boards
def interactive_evaluation():

    print("Interactive Tic Tac Toe Evaluation")
    print("Enter board using 'x', 'o', and '_' (underscore for empty)")

    board = []
    for i in range(3):
        while True:
            row_input = input(f"Enter row {i+1} (3 characters, e.g., 'x_o'): ").strip().lower()
            if len(row_input) == 3 and all(c in 'xo_' for c in row_input):
                board.append(list(row_input))
                break
            else:
                print("Invalid input! Please enter exactly 3 characters using only 'x', 'o', or '_'")

    print("\nYour Board:")
    print_board(board)

    score = evaluate_tic_tac_toe(board)
    print(f"Evaluation Score: {score}")

    if score == 10:
        print("Result: X wins!")
    elif score == -10:
        print("Result: O wins!")
    else:
        print("Result: Draw or game in progress")

# Main execution
if __name__ == "__main__":
    print("Tic Tac Toe Evaluation Function")
    print("=" * 40)

    # Run predefined test cases
    test_evaluation_function()

```
